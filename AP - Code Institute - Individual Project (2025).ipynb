```python
# 5.1 EXTRACT ‚Äî Load raw dataset

RAW_PATH = "highest_grossing_mobile_games.csv"

print(f"Attempting to load data from: {RAW_PATH}")

df_raw = pd.read_csv(RAW_PATH)

print("‚úîÔ∏è Raw data loaded successfully.")
print("Shape (rows, columns):", df_raw.shape)
print("\nFirst 5 rows:")
df_raw.head()
```

```python
# 5.2.1 Create a working copy
df = df_raw.copy()
print("Initial dataset shape:", df.shape)
df.head()

```

```python
# 5.2.2 Standardise column names

# After you have done df.columns = ... lowercased etc.

print("Columns before rename:", df.columns)

# Rename mislabelled column: 'global_sales' actually stores genre tags
df.rename(columns={'global_sales': 'genre_tags'}, inplace=True)

print("Columns after rename:", df.columns)


```

```python
# Inspect raw column before any cleaning

print("Raw columns in df_raw:\n", df_raw.columns, "\n")

print("First 20 raw values in Global_Sales (from df_raw):")
print(df_raw['Global_Sales'].head(20))

print("\nNon-null raw values sample:")
print(df_raw['Global_Sales'].dropna().head(20))
```

```python
# Re-create df from the cleaned data if needed
# (skip these two lines if df is already your cleaned working copy)
# df = df_raw.copy()

# Check what columns you have right now
print(df.columns)

# Create genre_tags from the Global_Sales column
# If your column is lowercase 'global_sales', use that name instead.
if 'Global_Sales' in df.columns:
    df['genre_tags'] = df['Global_Sales'].astype(str)
elif 'global_sales' in df.columns:
    df['genre_tags'] = df['global_sales'].astype(str)
else:
    raise ValueError("I can't find a Global_Sales / global_sales column.")
```

```python
# 5.2.3 Feature engineering from genre_tags

# 1) Primary genre (first tag before '/')
df['primary_genre'] = (
    df['genre_tags']
        .astype(str)
        .str.split('/')
        .str[0]
        .str.strip()
)

# 2) Number of genres assigned to each game
df['genre_count'] = (
    df['genre_tags']
        .astype(str)
        .str.split('/')
        .apply(len)
)

# Quick check
df[['genre_tags', 'primary_genre', 'genre_count']].head(10)
```

```python
# 5.2.4 Fix mislabelled columns and extract release year

# In this dataset, the 'year' column actually holds publisher names,
# and the 'publisher' column holds the release date.
# We rename them to reflect their true meaning.

df = df.rename(columns={
    'year': 'publisher_name',
    'publisher': 'release_date'
})

# Convert release_date to datetime
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

# Create a numeric release_year column
df['release_year'] = df['release_date'].dt.year

df[['publisher_name', 'release_date', 'release_year']].head()
```

```python
# 5.3 VALIDATE ‚Äì Quick missing values overview

missing = df.isna().sum().sort_values(ascending=False)
missing
```

```python
# 5.4 LOAD ‚Äì Save cleaned dataset to disk

CLEAN_PATH = "highest_grossing_mobile_games_cleaned.csv"

df.to_csv(CLEAN_PATH, index=False)

print(f"‚úîÔ∏è Cleaned dataset saved to: {CLEAN_PATH}")
print("Final cleaned shape:", df.shape)
```

### 7.1 Problem Definition

The prediction problem can be framed as a **regression task**:

> Given information about a game (year, genre, publisher, etc.),

This clearly-defined target allows us to:

- Evaluate models using standard regression metrics (MAE, RMSE, R¬≤),
- Compare simple baselines against more advanced models,
- Decide whether additional data is needed.

---

### 7.2 Baseline Approach (Conceptual)

Before building complex models, a project would normally start with one or more
**baselines**, for example:

- A **mean baseline** that predicts the average Global_Sales for every game,
- A **median baseline** that predicts the median Global_Sales,
- A simple lookup baseline that predicts the average sales per genre or per
  publisher.

These baselines require minimal code but give a clear **reference point**:
any machine learning model must perform better than these simple rules to be
considered useful.

---

### 7.3 Full Modelling Roadmap (Future Work)

4. **Model Interpretation**
   - Examine feature importances to see which variables drive predictions:
     - Are genres or publishers more important?

5. **Deployment Considerations**

# 9. Limitations & Considerations

While the analysis provides strong descriptive and strategic insights, there are
several limitations to acknowledge. These do not weaken the value of the work,
but instead demonstrate awareness of what additional data or methods would be
needed for a fully production-ready model.

---

## 9.1 Dataset Scope

The dataset reflects a **subset of mobile and multiplatform titles**, rather than
the entire global gaming market. As such:

- Sales figures may be incomplete or capped,
- Platform coverage may favour mobile ecosystems,
- Some genres appear over- or under-represented relative to real markets.

This means conclusions are accurate *within this dataset*, but may not
immediately generalise to external datasets without validation.

---

## 9.2 Missing Predictive Features

Although the dataset includes essential descriptors (genre, publisher, year),
several powerful predictors are not available:

- **Marketing budget**
- **User acquisition spend**
- **Game quality metrics** (ratings, reviews)
- **Franchise size**
- **LiveOps investment**
- **Team size**
- **Monetisation strategy** (premium vs freemium vs hybrid)

These variables are known to influence commercial outcomes, and their absence
limits the performance ceiling of any predictive model built on this dataset.

---

## 9.3 Categorical Complexity

Genres and publishers exhibit high cardinality:

- Many genres include multiple sub-tags,
- Publishers often appear only once (the ‚Äúlong tail‚Äù problem),
- Some entries contain overlapping category semantics.

This increases noise and makes encoding more challenging.

While this notebook avoids those complexities to focus on EDA quality, a more
advanced system would require:

- hierarchical genre clusters,
- publisher grouping,
- embedding-style encoders.

---

## 9.4 Temporal Bias

The dataset includes games from multiple years, but:

- Some years are heavily represented,
- Others contain only sparse data,
- Major industry transitions (e.g., Unity adoption, App Store algorithm changes)
  are not explicitly captured.

This means models would need careful handling of:

- market shifts,
- platform changes,
- lifecycle effects.

---

## 9.5 Non-Stationary Market Behaviour

Gaming markets evolve quickly. Patterns that hold in 2015 may not hold in 2025.

For example:

- Casual games exploded early but plateaued,
- Hypercasual rose then collapsed,
- Shooter and MOBA dominance varies by region,
- Monetisation trends shift globally.

This non-stationarity limits the longevity of any sales prediction model unless
it is updated regularly.

---

## 9.6 Methodological Constraints

This notebook intentionally:

- avoids complex modelling pipelines,
- avoids hyperparameter tuning,
- avoids advanced encoding systems,
- uses only clean EDA visualisation.

This keeps the scope focused, but also means:

> The analysis is *not* a full production ML project ‚Äî  
> it is a structured, insightful exploration designed for clarity and learning.

---

## 9.7 Summary of Limitations

| Category | Summary |
|---------|---------|
| Data Coverage | Not a complete market dataset |
| Missing Features | No marketing, budget, or quality metrics |
| Categorical Issues | High cardinality and overlapping labels |
| Temporal Effects | Market changes not explicitly modelled |

## 10.1 Enhanced Feature Engineering

- Adding region indicators (Asia, Europe, North America).
